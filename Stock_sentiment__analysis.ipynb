```python
data=pd.read_csv('/content/drive/My Drive/Data.csv',encoding="ISO-8859-1")
data.head()
```

```python
train=data[data['Date']<'20150101']
# print(train.head())
# print("**************************************************************")
test=data[data['Date']>'20141232']
# print(test.head())
```

```python
#punctuation removal
 
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]",' ',regex=True,inplace=True)

# data
```

```python
#renaming columns
lst=[i for i in range(25)]
new_index=[str(i) for i in lst]
data.columns=new_index
```

```python
data.head()
```

```python
# converting to lower case
for i in new_index:
    data[i]=data[i].str.lower()
data.head(1)
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headline=[]
for i in range(0,len(data.index)):
    headline.append(" ".join(str(x) for x in data.iloc[i,0:25]))
```

```python
 headline[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
#implement bag of words
CountVector=CountVectorizer(ngram_range=(2,2))
traindata=CountVector.fit_transform(headline)
```

```python
rfc=RandomForestClassifier(n_estimators=200,criterion='entropy')
rfc.fit(traindata,train['Label'])
```

```python
## predict for test data
test_trans=[]
for i in range(0,len(test.index)):
  test_trans.append(" ".join(str(x) for x in test.iloc[i,2:27]))
test_data=CountVector.transform(test_trans)
predictions=rfc.predict(test_data)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
```

```python
score=accuracy_score(test['Label'],predictions)
score
```

```python
report=classification_report(test['Label'],predictions)
print(report)
```
